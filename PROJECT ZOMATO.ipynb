# Solution 1.3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
df=pd.read_csv('D:\python\zomato.csv',encoding="latin-1")
df=df[df['Country Code']==1]
delhi=df[(df['City']=='New Delhi') | (df['City']=='Noida') | (df['City']=='Ghaziabad') | (df['City']=='Gurgaon') | (df['City']=='Faridabad')]
rest=df[(df['City']!='Faridabad') & (df['City']!='Ghaziabad') & (df['City']!='Gurgaon') & (df['City']!='New Delhi') & (df['City']!='Noida')]
data=delhi['Cuisines'].str.split(',')
data1=rest['Cuisines'].str.split(',')
d={}
for i in data:
    for row in i:
        key=row.strip()
        if key in d:
            d[key]+=1
        else:
            d[key]=1
sort_d = sorted(d.items(), key=lambda x: x[1], reverse=True)[:10]
#print(sort_d)
r={}
for i in data1:
    for row in i:
        key=row.strip()
        if key in r:
            r[key]+=1
        else:
            r[key]=1
sort_r = sorted(r.items(), key=lambda x: x[1], reverse=True)[:10]
#print(sort_r)
ncr=[]
ncr_value=[]
rest1=[]
rest_value=[]
print('Top 10 cuisines present in DELHI NCR are as follows ')
for i in range(len(sort_d)):
    print(sort_d[i][0],'->',sort_d[i][1])
    ncr.append(sort_d[i][0])
    ncr_value.append(sort_d[i][1])
plt.subplots(figsize=(16,6))
plt.title('TOP 10 CUISINES SERVED IN DELHI NCR')
plt.bar(ncr,ncr_value)
plt.show()
print('--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print('Top 10 cuisines present in REST-OF-INDIA are as follows ')
for i in range(len(sort_r)):
    print(sort_r[i][0],'->',sort_r[i][1])
    rest1.append(sort_r[i][0])
    rest_value.append(sort_r[i][1])
plt.subplots(figsize=(16,6))
plt.title('TOP 10 CUISINES SERVED IN REST-OF-INDIA')
plt.bar(rest1,rest_value)
plt.show()
```

# Solution 2.1.4

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
df=pd.read_csv('D:\python\zomato.csv',encoding="latin-1")
df=df[df['Country Code']==1]
df['Cuisines'].value_counts()
d={}
for ele in df['Cuisines']:
    li=[x.strip() for x in ele.split(',')]
    for v in li:
        d[v]=d.get(v,0)+1
        
sort_d=sorted(d.items(),key= lambda kv :(kv[1],kv[0]),reverse=True)[:15]

i=0
cuisines=[]

print('---------------------------------------------------------------------------------------------------------------')
print('TOP 15 CUISINES SERVED IN MOST RESTAURANTS OF INDIA ')

for i in range(len(sort_d)):
    cuisines.append(sort_d[i][0])
    print(sort_d[i][0],sort_d[i][1])
print('---------------------------------------------------------------------------------------------------------------')
  
l=list(zip(df['Cuisines'],df['Aggregate rating']))
d={}
for curr in cuisines:
    sum=0
    count=0
    for i in range(len(l)):
        if curr in l[i][0]:
            sum+=l[i][1]
            count+=1
    d[curr]=(sum/count)
x=[]
y=[]
for curr in d:
    x.append(curr)
    y.append(d[curr])
plt.subplots(figsize=(20,10))
plt.title('RATING VERSES CUISINES SERVED IN INDIA')
plt.xlabel('Cuisine served')
plt.ylabel('Rating')
plt.plot(x,y,color='red')
plt.show()
```

# Solution 2.2.1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
df=pd.read_csv('D:\python\zomato.csv',encoding="latin-1")
df=df[df['Country Code']==1]
#df=df.copy()
df['weighted_average']=df['Votes']*df['Aggregate rating']
l=list(zip(df['Locality'],df['weighted_average']))
s=df['Votes'].sum()
d={}
ind=[ele[0] for ele in l]
for val in ind:
    key=df[df['Locality']==val]['weighted_average'].sum()
    v=df[df['Locality']==val]['Votes'].sum()
    if v==0:
        continue
    d[val]=(key/v)

d=sorted(d.items(),key= lambda kv:(kv[1],kv[0]),reverse=True)[:10]
print('---------------------------------------------------------------------------------------------')
print('                          TOP 10 LOCALITIES HAVING BEST RATING                               ')
print()
x=[]
y=[]
for i in range(len(d)):
    x.append(d[i][0])
    y.append(d[i][1])
    print(d[i][0],'----->>>>',d[i][1])
print('---------------------------------------------------------------------------------------------')


plt.subplots(figsize=(30,10))
plt.plot(x,y)
plt.xticks(rotation=45)
plt.show()
```
